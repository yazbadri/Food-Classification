# Modeling with Neural Nets & Pre-trained Weights

## Modeling with Xception Pretrained Model

```python
# code help: https://keras.io/api/applications/
# code help: https://www.tensorflow.org/api_docs/python/tf/keras/applications/xception/Xception

# create the base pre-trained model
X_model = Xception(weights='imagenet', include_top=False)

# add a global spatial average pooling layer
x = X_model.output
x = GlobalAveragePooling2D()(x)
# let's add a fully-connected layer
x = Dense(256, activation='relu')(x)
# and a logistic layer 
predictions = Dense(3, activation='softmax')(x)

# this is the model we will train
model = Model(inputs=X_model.input, outputs=predictions)

# first: train only the top layers (which were randomly initialized)
# i.e. freeze all convolutional InceptionV3 layers
for layer in X_model.layers:
    layer.trainable = False

# compile the model (should be done *after* setting layers to non-trainable)
model.compile(optimizer=SGD(learning_rate=0.2, momentum=0.9), loss='categorical_crossentropy', metrics=['accuracy'])
early_stop = EarlyStopping(mode='auto', patience=5)

# train the model on the new data for a few epochs
X_results = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=30, batch_size=32, callbacks=early_stop, workers=4)

# let's visualize layer names and layer indices to see how many layers
# we should freeze:
for i, layer in enumerate(X_model.layers):
   print(i, layer.name)
```

```python
# plotting our accuracy for training and testing
plt.figure(figsize=(12,8))
plt.plot(X_results.history['accuracy'], label='Training Accuracy', color='blue')
plt.plot(X_results.history['val_accuracy'], label='Validation Accuracy', color='fuchsia')
plt.legend(fontsize='xx-large')
plt.title('Convolutional Xception Model Accuracy', size=28)
plt.xlabel('Epoch', size=20)
plt.ylabel('Accuracy', size=20)
plt.savefig('vgg_accuracy_3.png', bbox_inches='tight');
```

```python
# evaluating on test data
X_score = model.evaluate(X_test, y_test, verbose=1)

X_labels = model.metrics_names

# printing Xception scores
print(f'Xception {X_labels[0]}  : {X_score[0]}')
print(f'Xception {X_labels[1]}   : {X_score[1]}')
```

** Interpretation **
The Xception pre=trained model is doing pretty well on this subset, with 93-96% accuracy on testing data (100% on training). We'll try a VGG16 model below.

## Modeling with VGG16 Pretrained Model

```python
# code from https://www.analyticsvidhya.com/blog/2020/08/top-4-pre-trained-models-for-image-classification-with-python-code/

# instantiating VGG16 model
base_model = VGG16(input_shape = input_shape, # Shape of our images
include_top = False, # Leave out the last fully connected layer
weights = 'imagenet')

# freezing layers
for layer in base_model.layers:
    layer.trainable = False

# flattening the output layer to 1 dimension
x = layers.Flatten()(base_model.output)

# adding a fully connected layer
x = layers.Dense(32, activation='relu')(x)

# adding dropout to minimize overfitting
x = layers.Dropout(0.5)(x)

# adding a final layer with our number of classes for nodes
x = layers.Dense(3, activation='softmax')(x)

# instantiating the model we'll be training
vgg16 = tf.keras.models.Model(base_model.input, x)

# compiling our model
vgg16.compile(optimizer = tf.keras.optimizers.RMSprop(learning_rate=0.01), loss = 'categorical_crossentropy',metrics = ['accuracy'])

# early_stop = EarlyStopping(mode='auto', patience=5)

# train the model on the new data for a few epochs
vgg_results_3 = vgg16.fit(X_train, y_train, validation_data = (X_test, y_test), batch_size=32, epochs = 20)
```

```python
# Evaluate model on test data.
vgg_score = vgg16.evaluate(X_test, y_test, verbose=1)

vgg_labels = vgg16.metrics_names

# Compare Xception and VGG16 models.
print(f'Xception {X_labels[0]}  : {X_score[0]}')
print(f'Xception {X_labels[1]}   : {X_score[1]}')
print()
print(f'VGG16 {vgg_labels[0]}  : {vgg_score[0]}')
print(f'VGG16 {vgg_labels[1]}   : {vgg_score[1]}')
```

## Modeling with Convolutional Neural Nets

```python
# code from lesson 8.03
cnn_model = Sequential()

cnn_model.add(Conv2D(filters=4, # number of filters
                     kernel_size=(3,3),# height/width of filter
                     activation='relu', # activation function
                     input_shape=(input_shape)))

cnn_model.add(MaxPooling2D(pool_size=(2, 2)))

cnn_model.add(Conv2D(filters=6,
                     kernel_size=(3,3),
                     activation='relu'))

cnn_model.add(MaxPooling2D(pool_size=(2,2)))

cnn_model.add(Conv2D(filters=6,
                     kernel_size=(3,3),
                     activation='relu'))

cnn_model.add(MaxPooling2D(pool_size=(2,2)))


cnn_model.add(Flatten())
# cnn_model.add(Dense(512, activation='relu'))
cnn_model.add(Dense(256, activation='relu'))
cnn_model.add(Dropout(0.2))
cnn_model.add(Dense(128, activation='relu'))
cnn_model.add(Dropout(0.1))
cnn_model.add(Dense(128, activation='relu'))
cnn_model.add(Dropout(0.05))
cnn_model.add(Dense(3, activation='softmax'))

cnn_model.compile(loss='categorical_crossentropy',
                  optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
                  metrics=['accuracy'])

early_stop = EarlyStopping(patience = 10)
```

```python

cnn_results = cnn_model.fit(X_train, y_train,
                            batch_size=32,
                            validation_data=(X_test, y_test),
                            epochs=20,
                            callbacks=early_stop,
                            verbose=1,
                            workers=4)
```

```python
cnn_model.summary()
```

```python
# Check out our train loss and test loss over epochs.
train_loss = cnn_results.history['loss']
test_loss = cnn_results.history['val_loss']

# Set figure size.
plt.figure(figsize=(12, 8))

# Generate line plot of training, testing loss over epochs.
plt.plot(train_loss, label='Training Loss', color='#185fad')
plt.plot(test_loss, label='Testing Loss', color='orange')

# Set title
plt.title('Training and Testing Loss by Epoch', fontsize=25)
plt.xlabel('Epoch', fontsize=18)
plt.ylabel('Categorical Crossentropy', fontsize=18)

plt.legend(fontsize = 18);
```

```python
# Check out our train loss and test loss over epochs.
train_acc = cnn_results.history['accuracy']
test_acc = cnn_results.history['val_accuracy']

# Set figure size.
plt.figure(figsize=(12, 8))

# Generate line plot of training, testing loss over epochs.
plt.plot(train_acc, label='Training Accuracy', color='#185fad')
plt.plot(test_acc, label='Testing Accuracy', color='orange')

# Set title
plt.title('Training and Testing Accuracy by Epoch', fontsize=25)
plt.xlabel('Epoch', fontsize=18)
plt.ylabel('Categorical Crossentropy', fontsize=18)

plt.legend(fontsize = 18);
```

```python
# Evaluate model on test data.
cnn_score = cnn_model.evaluate(X_test, y_test, verbose=1)

cnn_labels = cnn_model.metrics_names

# Compare CNN, Xception and VGG16 models
print(f'CNN {cnn_labels[0]}  : {cnn_score[0]}')
print(f'CNN {cnn_labels[1]}   : {cnn_score[1]}')
print()
print(f'Xception {X_labels[0]}  : {X_score[0]}')
print(f'Xception {X_labels[1]}   : {X_score[1]}')
print()
print(f'VGG16 {vgg_labels[0]}  : {vgg_score[0]}')
print(f'VGG16 {vgg_labels[1]}   : {vgg_score[1]}')
```

### ** Interpretation **

The CNN model, with 3 convolutional layers, 4 dense layers and 3 dropout layers, performed much worse than our VGG16 and Xception models.
